```python
import pandas as pd 
import xlsxwriter as ex
import numpy as np
import scipy
from scipy.stats import pearsonr
import os
```

```python
wiki_data = pd.read_csv('normalized_disease_wiki_0_6_range.csv')

for name_of_file in os.listdir('/home/bbardak/Desktop/Tez/MultipleDisease/sonson/correlations/combined/wiki_normalized/'):
    print name_of_file[:10]
    if name_of_file[-3:] == 'csv' and name_of_file[:10] != 'normalized':
        train_data = pd.read_csv(name_of_file, header=None)
        workbook = ex.Workbook('pearson_between'+name_of_file+'.xlsx')
        worksheet = workbook.add_worksheet()
        col = 0
        row = 0
        for i in wiki_data:
            row = 0
            for j in train_data:
                if i != 'Unnamed: 0' and j != 0:
                    print i, j, name_of_file
                    worksheet.write(row,col,pearsonr(wiki_data[i],train_data[j])[0])
                    row = row + 1
            col = col + 1
        workbook.close()
```
